```python
# 调用各种包

import pandas as pd

# import seaborn, but current version of seaborn generates a bunch of warnings that we will ignore

import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)

# load the dataset
# 需要注意的是，notebook中引入数据的时候的斜杠方向和系统自带的斜杠方向是相反的

iris = pd.read_csv("C:/Users/XPS/Kaggle/Iris/Iris.csv")

iris.head()
```

```python
# 看一下每一个种类的个数

iris["Species"].value_counts()
```

```python
# 根据pandas中的.plot命令绘制散点图

iris.plot(kind="scatter" , x="SepalLengthCm" , y="SepalWidthCm")
```

```python
# 也可以用seaborn来作图
# a seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure
# 直方图可以直观的显示出这个值的样本数量

# size用来调节图片的大小

sns.jointplot(x="SepalLengthCm" , y="SepalWidthCm" , data = iris , size =5)
```

```python
# 但是上图无法分辨出每个点分别属于哪个类别
# 使用 seaborn's FacetGrid to color the scatterplot by species

g = sns.FacetGrid(iris , hue = "Species" , size =5) 
g.map(plt.scatter, "SepalLengthCm" , "SepalWidthCm") 
g.add_legend()
```

```python
# 上面这种代码的另外一种写法


g = sns.FacetGrid(iris , hue = "Species" , size =5) \
       .map(plt.scatter, "SepalLengthCm" , "SepalWidthCm") \
       .add_legend()
```

```python
# 通过seaborn的boxplot观察每个分类单个特征的差异性

sns.boxplot(x="Species", y="PetalLengthCm" , data = iris )
```

```python
# extend the plot by adding a layer of individual points on top of the previous plot
# jitter = True so that points don't fall in single vertical lines
# saving the resulting axes as ax each time to present them at the same time

ax = sns.boxplot(x="Species", y="PetalLengthCm" , data = iris )
ax = sns.stripplot(x="Species", y="PetalLengthCm" , data = iris, jitter = True , edgecolor = "gray")
```

```python
# extend the plot by adding a layer of individual points on top of the previous plot
# jitter = False so that points falls in single vertical lines
# saving the resulting axes as ax each time to present them at the same time

ax = sns.boxplot(x="Species", y="PetalLengthCm" , data = iris )
ax = sns.stripplot(x="Species", y="PetalLengthCm" , data = iris, jitter = False , edgecolor = "gray")
```

```python
# 将boxplot和stripplot的优点结合起来——violin plot

sns.violinplot(x="Species", y="PetalLengthCm" ,data = iris , size=6)
```

```python
# kdeplot, which creats and visualizes a kernel density estimate of the underlying feature

g = sns.FacetGrid(iris, hue="Species" , size = 6)
g.map(sns.kdeplot , "PetalLengthCm")
g.add_legend()
```

```python
# pairplot shows the bivariate relation between each pair of features
# from pairplot, we will see that the Iris-setosa species (蓝色的部分)is seperated from the other

sns.pairplot(iris.drop("Id",axis=1), hue="Species", size=3)
```

```python
sns.pairplot(iris.drop("Id",axis=1), hue="Species", size=3, diag_kind = "kde")
```

```python
# 仅用pandas也可以绘制箱线图

iris.drop("Id", axis=1).boxplot(by="Species",figsize=(12,6))
```
